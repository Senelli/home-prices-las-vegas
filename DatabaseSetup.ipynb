```python
import psycopg2
from psycopg2 import sql, extras
import csv
from sqlconfig import password as password

cities_zip_codes_path = "Resources/cities_zip_codes.csv"
las_vegas_home_values_path = "Resources/las_vegas_home_values_by_zip_codes.csv"
```

```python
NEON_CONN = f"postgresql://neondb_owner:{password}@ep-late-bonus-apm89qdo.c-7.us-east-1.aws.neon.tech/neondb?sslmode=require"

conn = psycopg2.connect(NEON_CONN)
cursor = conn.cursor()

# Insert city_zipcodes
with open(cities_zip_codes_path, 'r') as f:
    reader = csv.reader(f)
    next(reader)
    extras.execute_values(cursor, "INSERT INTO city_zipcodes VALUES %s ON CONFLICT DO NOTHING", reader)

# Insert home_values
with open(las_vegas_home_values_path, 'r') as f:
    reader = csv.reader(f)
    next(reader)
    extras.execute_values(cursor, "INSERT INTO home_values VALUES %s", reader)

conn.commit()
cursor.close()
conn.close()
print("Done! Data uploaded to Neon.")
```
